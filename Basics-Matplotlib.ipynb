# Matplotlib Library

## Pyplot Basics

#### Objects Hierarchy:
* Figure
    * Axes
        * x-axis
        * y-axis
        * spines
        * legends

#### Components: 
* Spines - Lines connecting the axis tick marks (Border lines of the graph)
* Title - Text label of the whole Figure object
* Legend - Describe the content of the plot
* Grid - Vertical and horizontal lines used as an extension of the tick marks
* X/Y axis label - Text labels for the X and Y axes below the spines
* Minor tick - Small value indicators between the ajor tick marks
* Minor tick label - Text label that will be displayed at the minor ticks

Check out the Anatomy of a Figure object: https://matplotlib.org/3.1.1/gallery/showcase/anatomy.html

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

### Line Chart

#### Plot a Line chart

```python
# Basic plot a line

# Create a figure object with number 5, dpi, width & height
plt.figure(figsize=(5, 3), num=5, dpi=125)

# Find the current figure object number
display(plt.gcf().number)

# Plot a line
plt.plot([0, 1, 2, 3], [2, 4, 6, 8], 'o')

# Display
plt.show()

# Close the figure object of specified figure number
plt.close(5)
```

#### Plot a Multiline chart

```python
# Basic plot 2 series of lines

# Create a figure object with number 5, dpi, width & height
plt.figure(figsize=(5, 3), dpi=125)

# Setup data
line1 = [2, 4, 6, 8]
line2 = [1, 5, 9, 13]

# Plot 2 line series
plt.plot(line1, '-s', line2 , '-o')

# Display
plt.show()

# Close the figure object of specified figure number
plt.close()
```

#### Plot a Line chart with customizations

```python
# Basic plot 2 series of lines

# Create a figure object with number 5, dpi, width & height
plt.figure(figsize=(5, 3), dpi=125)

# Setup data
line1 = [2, 4, 6, 8]
line2 = [1, 5, 9, 13]

# Plot each line with custom params
plt.plot(line1, color='blue', label='Label 1', marker='s', linestyle='dashed', linewidth=2, markersize=12)
plt.plot(line2, color='green', label='Label 2', marker='o', linestyle='dashed', linewidth=2, markersize=12)

# Customize the ticks (a.k.a change the graph indices)
plt.xticks(ticks=np.arange(6), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'], rotation=20)
plt.yticks(ticks=np.arange(0, 16, 2))

# Add legend
plt.legend()

# Add a chart title
# fig = plt.figure()
# fig.suptitle('suptitle', fontsize=10, fontweight='bold')
plt.title('Customized Multi line chart sample', fontsize=16)

# Add labels
ax = plt.gca()
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')

# Add a grid
ax.grid()

# Add a text in the plot
ax.text(3, 4, 'Text in Data', bbox={'facecolor': 'blue', 'alpha': 0.5, 'pad': 5})

# Add a annotation in the plot
ax.annotate('Annotate text', xy=(4,2), xytext=(6,4), arrowprops=dict(facecolor='green', shrink=0.05))

# Save the chart to file with thin outer white margin
plt.savefig('lineplot.png', dpi=200, bbox_inches='tight')

# Read the image from saved file and display it
plt.imread('lineplot.png')
plt.show()

# Close the figure object of specified figure number
plt.close()
```
